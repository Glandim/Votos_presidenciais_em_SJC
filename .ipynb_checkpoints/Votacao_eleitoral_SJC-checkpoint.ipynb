```python
#Importamos as bibliotecas
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sd
low_memory=False
%matplotlib inline
pd.options.display.max_columns = 90
pd.options.display.max_rows = 90
```

```python
#Foi ultizada a variável "df" para receber o arquivo lido 
df = pd.read_csv(r'dataBase1Turno.csv', delimiter=',',  encoding='latin-1')
```

```python
#O dicionario "candidatos" tem a função de de armazenar todos os nomes do candidatos como Key
#e como Value é atratribuido um valor fixo de 0
candidatos = {}
dataPres = df["NM_VOTAVEL"]
for x in dataPres:
    if x not in candidatos: candidatos[x] = 0
candidatos
```

```python
#O dicionario "zonas" recebe todas as zonas do documento como Keys e como Value 
#recebe o conteudo do dicionario "candidatos"
zonas = {}
dataZonas = df["NR_ZONA"]
for x in dataZonas:
    if x not in zonas: zonas[x] = {'Branco': 0, 'Nulo': 0, 'CIRO GOMES': 0, 'FERNANDO HADDAD': 0, 'HENRIQUE MEIRELLES': 0, 'VERA': 0, 'JAIR BOLSONARO': 0, 'MARINA SILVA': 0, 'ALVARO DIAS': 0, 'JOAO AMOEDO': 0, 'GERALDO ALCKMIN': 0, 'GUILHERME BOULOS': 0, 'CABO DACIOLO': 0, 'EYMAEL': 0, 'JOAO GOULART FILHO': 0}
zonas        
```

```python
#Os valores que estvam em 0 são alterados para a soma de total dos votos por candidatos
lenRows = len(df["NM_VOTAVEL"])
for row in range(0, lenRows):
    zonas[df["NR_ZONA"][row]][df["NM_VOTAVEL"][row]] += df["QT_VOTOS"][row]
zonas
```

```python
#Pegamos do dicionario "zonas" os nomes dos candidatos e os valores
#e colocamos respectivamente nas variáveis "names" e "values"
names = list(zonas[127].keys())
values = list(zonas[127].values())

#Criamos o gráfico 
fig = plt.figure()
ax = fig.add_axes([0,0,3.5,2.5])
ax.bar(names,values)

#criamos as legendas
ax.set_title("Zona eleitoral 127", fontsize=(20));
ax.set_ylabel("Votos", fontsize=(15));
ax.set_xlabel("Candidatos", fontsize=(15));

#Apresentamos o gráfico
plt.show()

#por conta da quantidade de candidatos, o gráfico tem uma escala que dificulta a leitura dos nomes.
#Uma solução que decidimos trazer é apresentar um gráfico com apenas os 5 candidatos mais votados por zona,
#que será apresentado a partir do próximo gráfico
```

```python
#Organizamos do menos votado ao mais votado
sort = sorted(zonas[127].items(), key=lambda kv: kv[1])

#Pegamos somente os 5 mais votados
sort2 = sort[-5::]

#Separamos o nome e os total de votos nas variáveis "keys" e "values" para criação do gráfico
keys = []
values = []
for x in sort2:
    keys.append(x[0])
    values.append(x[1])
    
#Mostramos a tabela do menos votado ao mais votado dessa zona
print(pd.DataFrame(sort, columns=["Candidatos", "Votos"]))
```

```python
#Criamos o gráfico 
fig = plt.figure()
ax = fig.add_axes([0,0,2,2])
ax.bar(keys,values)

#criamos as legendas
ax.set_title("Zona eleitoral 127", fontsize=(20));
ax.set_ylabel("Votos", fontsize=(15));
ax.set_xlabel("Candidatos", fontsize=(15));

#Apresentamos o gráfico
plt.show()
```
